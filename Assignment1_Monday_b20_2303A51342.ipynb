```python
T=int(input())
for i in range(T):
    N=int(input())
    meetings=[]
    for i in range(N):
        Si,Ei=map(int,input().split())
        meetings.append((Si,Ei))
    meetings.sort(key=lambda x: x[1])
    max_meetings=0
    last_end=0
    for Si,Ei in meetings:
        if Si>=last_end:
            max_meetings=max_meetings + 1
            last_end=Ei
    print(max_meetings)
```
