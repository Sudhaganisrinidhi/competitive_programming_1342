```python
class SegmentTree:
    def __init__(self, arr):
        self.n = len(arr)
        self.arr = arr
        self.min_tree = [0] * (4 * self.n)
        self.max_tree = [0] * (4 * self.n)
        self.build(0, 0, self.n - 1)

    def build(self, node, start, end):
        if start == end:
            self.min_tree[node] = self.arr[start]
            self.max_tree[node] = self.arr[start]
        else:
            mid = (start + end) // 2
            self.build(2*node+1, start, mid)
            self.build(2*node+2, mid+1, end)
            self.min_tree[node] = min(self.min_tree[2*node+1],
                                      self.min_tree[2*node+2])
            self.max_tree[node] = max(self.max_tree[2*node+1],
                                      self.max_tree[2*node+2])
    def range_min(self, node, start, end, l, r):
        if r < start or end < l:
            return 10**9
        if l <= start and end <= r:
            return self.min_tree[node]
        mid = (start + end) // 2
        return min(
            self.range_min(2*node+1, start, mid, l, r),
            self.range_min(2*node+2, mid+1, end, l, r)
        )

    def range_max(self, node, start, end, l, r):
        if r < start or end < l:
            return -10**9
        if l <= start and end <= r:
            return self.max_tree[node]
        mid = (start + end) // 2
        return max(
            self.range_max(2*node+1, start, mid, l, r),
            self.range_max(2*node+2, mid+1, end, l, r)
        )
    def update(self, node, start, end, idx, value):
        if start == end:
            self.arr[idx] = value
            self.min_tree[node] = value
            self.max_tree[node] = value
        else:
            mid = (start + end) // 2
            if idx <= mid:
                self.update(2*node+1, start, mid, idx, value)
            else:
                self.update(2*node+2, mid+1, end, idx, value)

            self.min_tree[node] = min(self.min_tree[2*node+1],
                                      self.min_tree[2*node+2])
            self.max_tree[node] = max(self.max_tree[2*node+1],
                                      self.max_tree[2*node+2])
n = int(input())
arr = list(map(int, input().split()))
st = SegmentTree(arr)
q = int(input())
l, r = map(int, input().split())
print(st.range_min(0, 0, n-1, l, r))
l, r = map(int, input().split())
print(st.range_max(0, 0, n-1, l, r))
idx, val = map(int, input().split())
st.update(0, 0, n-1, idx, val)
l, r = map(int, input().split())
print(st.range_min(0, 0, n-1, l, r))
```
