```python
class DisjointSet:
    def __init__(self, n):
        self.parent = []
        for i in range(n):
            self.parent.append(i)
    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]
    def union(self, u, v):
        root_u = self.find(u)
        root_v = self.find(v)
        if root_u != root_v:
            self.parent[root_v] = root_u
    def connected(self, u, v):
        return self.find(u) == self.find(v)
n = int(input("Number of servers: "))
ds = DisjointSet(n)
m = int(input("Number of connections: "))
print("Enter connections:")
for _ in range(m):
    u, v = map(int, input().split())
    ds.union(u, v)
q = int(input("Number of connectivity queries: "))
print("Enter queries:")
for _ in range(q):
    u, v = map(int, input().split())
    if ds.connected(u, v):
        print("YES")
    else:
        print("NO")
```
