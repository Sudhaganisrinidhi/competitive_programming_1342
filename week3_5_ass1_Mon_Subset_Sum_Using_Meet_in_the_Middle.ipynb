```python
def subset_sum_mitm(arr, target):
    n = len(arr)
    mid = n // 2

    left = arr[:mid]
    right = arr[mid:]

    left_sums = []

    for mask in range(1 << len(left)):
        s = 0
        for i in range(len(left)):
            if mask & (1 << i):
                s += left[i]
        left_sums.append(s)

    right_sums = []

    for mask in range(1 << len(right)):
        s = 0
        for i in range(len(right)):
            if mask & (1 << i):
                s += right[i]
        right_sums.append(s)

    right_sums.sort()

    for x in left_sums:
        need = target - x
        l, r = 0, len(right_sums) - 1
        while l <= r:
            m = (l + r) // 2
            if right_sums[m] == need:
                return True
            elif right_sums[m] < need:
                l = m + 1
            else:
                r = m - 1

    return False


t = int(input())

for _ in range(t):
    n, s = map(int, input().split())
    arr = list(map(int, input().split()))

    if subset_sum_mitm(arr, s):
        print("YES")
    else:
        print("NO")
```
