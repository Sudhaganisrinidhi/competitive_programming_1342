```python
def subset_sum(arr, target, index, current_sum):
    if current_sum == target:
        return True
    if index == len(arr) or current_sum > target:
        return False
    if subset_sum(arr, target, index + 1, current_sum + arr[index]):
        return True
    if subset_sum(arr, target, index + 1, current_sum):
        return True
    return False
n = int(input())
arr = list(map(int, input().split()))
target = int(input())

result = subset_sum(arr, target, 0, 0)

if result:
    print("True")
else:
    print("False")
```
